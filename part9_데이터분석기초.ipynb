```python
import csv # 라이브러리 임포트
f = open('./data/card.csv', encoding='utf8') # 파일 열기
data = csv.reader(f) # 파일 읽기
next(data) # 헤더 제거
data=list(data) # 리스트로 변환
print(data)
```

```python
spend = 0
for row in data:
    # print(row[-1])
    if row[-1] == '전표매입':
        spend += int(row[-3])
print('총 이용금액: ',spend,'원')
```

```python
import matplotlib.pyplot as plt
s_mon = [0,0,0]
for row in data:
    if row[-1] == '전표매입':
        month = row[0].split()[0].split('/')[1]
        if month == '10':
            s_mon[0]+=int(row[6])
        elif month =='11':
            s_mon[1]+=int(row[6])
        elif month == '12':
            s_mon[2]+=int(row[6])
# print(s_mon)

plt.rc('font', family='Gulim')
plt.title('10월~12월 지출현황')
plt.bar(['10월','11월','12월'],s_mon,color='royalblue')
plt.show()
```

```python
taxi = [0,0,0]
for row in data:
    if row[-1] == '전표매입' and '택시' in row[5]:
        month = row[0].split()[0].split('/')[1]
        if month == '10':
            taxi[0]+=int(row[6])
        elif month =='11':
            taxi[1]+=int(row[6])
        elif month == '12':
            taxi[2]+=int(row[6])

print(taxi)
plt.rc('font', family='Gulim')
plt.title('10월~12월 택시비')
plt.plot(['10월','11월','12월'],taxi,color='crimson', label='택시비 지출액')
plt.legend()
plt.show()
```

```python
deli = [0,0,0]
for row in data:
    if row[-1] == '전표매입' and row[5] == '(주)우아한형제들':
        month = row[0].split()[0].split('/')[1]
        if month == '10':
            deli[0]+=int(row[6])
        elif month =='11':
            deli[1]+=int(row[6])
        elif month == '12':
            deli[2]+=int(row[6])

print(deli)
plt.rc('font', family='Gulim')
plt.title('10월~12월 배달비')
plt.plot(['10월','11월','12월'],deli,color='indigo', label='배달비 지출액')
plt.legend()
plt.show()
```

```python
plt.plot(['10월','11월','12월'],taxi,color='crimson', label='택시비 지출액')
plt.plot(['10월','11월','12월'],deli,color='indigo', label='배달비 지출액')
plt.title('10~1월 택시비/배달비 지출현황')
plt.legend()
plt.show()
```

```python
import operator # 밸류 값으로 정렬할때 필요한 라이브러리

spending = {}
for row in data:
    if row[-1] == '전표매입':
        store, price = row[-4], int(row[-3])
        if store not in spending.keys():
            spending[store] = price
        else:
            spending[store] += price
top10 = sorted(spending.items(), key=operator.itemgetter(1), reverse=True)[:10] # 밸류값 내림차순 정렬 후 상위 10개 리턴
# print(top10)

top10_store, top10_amount = [], []
for t in top10:
    top10_store.append(t[0])
    top10_amount.append(t[1])

plt.barh(top10_store,top10_amount)
plt.show()
```
